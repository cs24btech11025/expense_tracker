```python
import pandas as pd
df=pd.read_csv("Pictures/expenses.csv")
print(df.head())
print(df.shape)
```

```python
df["date"]=pd.to_datetime(df["date"])
df["month"]=df["date"].dt.strftime("%B")
print(df.dtypes)
print(df.isnull().sum())
```

```python
print("Total spent:", df['amount'].sum())
print("\nSpending by category:")
print(df.groupby('category')['amount'].sum().sort_values(ascending=False))
print("\nSpending by month:")
print(df.groupby('month')['amount'].sum())
```

```python
import matplotlib.pyplot as plt
category_spend=df.groupby("category")["amount"].sum()
plt.figure(figsize=(8,5))
category_spend.plot(kind="bar",color="black")
plt.title("spending by category")
plt.xlabel("category")
plt.ylabel("amount")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
category_spend.plot(kind="pie",autopct="%1.1f%%",startangle=0)
plt.title("expense breakdown")
plt.ylabel("")
plt.show()
```
